# Plotting manual assessment

```python
test = pd.read_excel("Evaluation/assessment lists.xlsx", engine='openpyxl')
test
```

```python
test.iloc[0,3:].values
```

```python
df = pd.DataFrame()
df["dirPrecision"] = [0.95 for x in test.columns[3:]]
df["dirRelevance"] = test.iloc[0,3:].values # invert the relevance of the direct linking, becasue 0 is most relevant
df["dirSTDPrecision"] = test.iloc[5,3:].values


df["TFPrecision"] = test.iloc[1,3:].values
df["TFRelevance"] = test.iloc[2,3:].values
df["TFSTDPrecision"] = test.iloc[7,3:].values

df["docPrecision"] = test.iloc[3,3:].values
df["docRelevance"] = test.iloc[4,3:].values
df["docSTDPrecision"] = test.iloc[9,3:].values
df = df.astype("float64")
df
```

```python
from numpy.polynomial.polynomial import polyfit

#plt.style.use('seaborn-whitegrid')
plt.style.use('ggplot')
#plt.style.use('science')

figure(figsize=(10, 5), dpi=160)

c1 = ("#348abd")
c2 = ("#e24a33")
c3 = ("#777777")



x = df["dirRelevance"].values
y = df["dirPrecision"].values
b, m = polyfit(x, y, 1)
#plt.plot(x, b + m * x, '--', c = c1)
#plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), c=c1, linestyle='dashed')

#plt.errorbar(df["dirRelevance"], df["dirPrecision"], yerr=df["dirSTDPrecision"], linestyle="None", c=c1,  capsize=3)
#plt.scatter(df["dirRelevance"], df["dirPrecision"], label = "Direct linking", c=c1)



x = df["TFRelevance"].values
y = df["TFPrecision"].values
b, m = polyfit(x, y, 1)
#plt.plot(x, b + m * x, '--', c = c2)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), c=c2, linestyle='dashed')

plt.errorbar(df["TFRelevance"], df["TFPrecision"], yerr=df["TFSTDPrecision"], linestyle="None", c=c2,  capsize=3)
plt.scatter(df["TFRelevance"], df["TFPrecision"], label = "TF-IDF", c=c2)




x = df["docRelevance"].values
y = df["docPrecision"].values
b, m = polyfit(x, y, 1)
#plt.plot(x, b + m * x, 'o', c = c3)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), c=c3, linestyle='dashed')

plt.errorbar(df["docRelevance"], df["docPrecision"], yerr=df["docSTDPrecision"], linestyle="None", c=c3,  capsize=3)
plt.scatter(df["docRelevance"], df["docPrecision"], label = "Doc2Vec", c=c3)

#plt.errorbar(df["dirRelevance"], df["dirPrecision"], yerr=df["dirSTDPrecision"], fmt='o', c=c1)

#plt.xlim(0, 1)
#plt.ylim(0, 1)

plt.xlabel("Average Relevance")
plt.ylabel("Precision")

plt.legend()

plt.savefig("Evaluation/Precision vs relevance scatter.png")
```
