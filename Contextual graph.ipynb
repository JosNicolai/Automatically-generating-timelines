# Creating a network
## Start with creating an edge list

```python
document_item_match = pd.read_csv("NewProcessingData/document_item_match.csv").set_index("Document_id")
document_item_match
```

```python
# count the total number of matches in the dataframe
total_matches = 0
for index, row in document_item_match.iloc[:,:].iterrows():
    for item in row:
        if pd.notnull(item):
            total_matches+=1
total_matches
```

```python
from itertools import combinations_with_replacement

# store the source and target items in a list
sources = []
targets = []

matches = []

# loop through every document
for i in tqdm(range(len(document_item_match.index))):
    # get all items in this row
    #items = [x for x in document_item_match.iloc[i].tolist() if pd.notnull(x)]
    items = [x for x in document_item_match.iloc[i].tolist() if pd.notnull(x)]
    
    # only loop through the list if there is more than one match
    if len(items) > 1:
        # create all possible combinations of length 2 out of the values includes identical combinations
        combinations = list(combinations_with_replacement(items, 2))
        #print(combinations)
        
        # loop through all possible combinations
        for comb in combinations:
            source = comb[0]
            target = comb[1]
            
            
            # and only add the combination if they are not equal to each other
            # and only if this match has not already been added
            if source != target and (source, target) not in matches:
                #print(source, target, source!=target)
                sources.append(source)
                targets.append(target)
                matches.append((source,target))
```

```python
# create a dataframe with every item id set out against every other item id
shortest_paths = pd.DataFrame()
shortest_paths["items"] = itemsInformation.keys()
shortest_paths = shortest_paths.set_index("items").T
shortest_paths["items"] = itemsInformation.keys()
shortest_paths = shortest_paths.set_index("items")
shortest_paths
```

# Calculate shortest paths between items

```python
# now add the shortest paths two the dataframe
# by looping through every match
for rowID in tqdm(itemsInformation):
    for columnID in itemsInformation:
        
        # if they are a match set distance to 0 and skip
        if rowID == columnID:
            shortest_paths.loc[rowID, columnID] = 0
            continue
        
        # else if they exist and have a path calculate the distance, otherwise set to -1
        else :
            if rowID in graph and columnID in graph:
                if nx.has_path(graph, rowID, columnID):
                     shortest_paths.loc[rowID, columnID] = nx.shortest_path_length(graph, rowID, columnID)
                else :
                    shortest_paths.loc[rowID, columnID] = np.nan
                
```
