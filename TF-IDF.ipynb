```python
# importing the necessary packages
import numpy as np
import pandas as pd
import os
import re
import pickle
from tqdm import tqdm

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

import spacy
# if necessary download the pipeline
#!python -m spacy download nl_core_news_sm

# setting the main file path to where all files are
os.chdir("C:\\Users\\SJNic\\Documents\\University\\Master\\Thesis\\Thesis Code")

# reading all lemmas. The converters is necessary in order to keep the lemmas as a list
#lemmas = pd.read_csv("Similarities/lemmas.csv", converters={'lemmas': eval}) 
#lemmas
lemmas = pd.read_pickle("similarities/newLemmas.pkl").to_frame()
#item_lemmas = pd.read_pickle("similarities/newLemmas.pkl")
#item_lemmas
lemmas
```

```python
# explanation
# https://stackoverflow.com/questions/12118720/python-tf-idf-cosine-to-find-document-similarity
```

```python
# because the strings have already been processed, the vectorizer does not need to do that itself
def dummy(doc):
    return doc

tfidf = TfidfVectorizer(
    tokenizer = dummy,
    preprocessor = dummy,
    lowercase = False
)

tfidf = tfidf.fit_transform(lemmas.lemmas)
```

```python
with open('Similarities/fitted_tfidf.pkl', 'wb') as fin:
    pickle.dump(tfidf, fin)
```

```python
TF_IDF = pd.DataFrame()
TF_IDF["doc"] = lemmas.index
TF_IDF = TF_IDF.set_index("doc").T
TF_IDF["doc"] = lemmas.index
TF_IDF = TF_IDF.set_index("doc").T
TF_IDF
```

```python
# loop through every document and calculate the similarity
# takes around 4 hours
for i in tqdm(range(len(lemmas.index))):
    key = lemmas.index[i]
    
    TF_IDF.loc[key] = linear_kernel(tfidf[i], tfidf).flatten()
TF_IDF
```

```python
TF_IDF.to_pickle("Similarities/new_TF_IDF.pkl")
```

# Find the most related items

```python
import pickle

with open("ProcessedMeetings/itemsInformation.txt", "rb") as fp:   
    itemsInformation = pickle.load(fp)
```

```python
# function for returning the most relevant items based on 
def return_related_items(df, id, min_value, print_bool = False):
    Related_items = df.loc[id][df.loc[id] > min_value].index
    
    if print_bool:
        print("The item given to this function:\n")    
        [print(x, itemsInformation[id][x]) for x in itemsInformation[id]]

        print("\nRelated items:")
        
        [print(f"{itemsInformation[x]['features']} {itemsInformation[x]['title']}") for x in Related_items]
    
    return Related_items
```

```python
def find_item(itemName, itemsInformation, print_bool = False):
    for key in itemsInformation:
        
        if itemsInformation[key]["title"] == itemName:
            
            for child in itemsInformation[key]:
                if print_bool:
                    print(f"{child}: \t\t{itemsInformation[key][child]}")
            if print_bool:
                print("_________________________________________")
            return itemsInformation[key]["item id"]

    print("No item related to this name could be found")
    return ""
```
